Import needed package for data analysis

```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

Data source: 

```python
data = pd.read_csv('data/edgar-co2-emissions.csv', index_col=0)
```

```python
data.head()
```

```python
unit = "kiloton"
```

```python
dataUSA = data.loc['USA', :]
dataUSA.head()
```

```python
fig1, ax1 = plt.subplots()

for key, grp in dataUSA.groupby(['Sector']):
    ax1 = grp.plot(ax=ax1, kind='line', x='Year', y='Emissions', label= key)
    plt.legend(loc = 'best')
    plt.xlabel('Year')
    plt.ylabel('Emissions (Kiloton)')
    plt.title('Emissions of CO2 from diffrent sectors in the U.S.')

plt.show()
```

```python
##Here's a comment from Ece: this looks great ye!
boxplot = dataUSA.boxplot('Emissions', by='Sector', grid = False, rot = 75)
plt.title("Box plot of CO2 emissions from different sectors in the U.S.")
plt.suptitle("")
plt.ylabel('Emissions (Kiloton)')

plt.show()
```
