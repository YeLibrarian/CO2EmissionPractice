### Which countries had GDP higher than the European average for that year?

```python
mask_higher = data > data.mean(axis=0)
mask_higher.head()
```

```python
def sumsq(x):
    return sum(x)*sum(x)
```

```python
data.columns
```

### Aggregate can take any callable function (like sumsq from above) in place of 'sum'

### The verbose `.aggregate(...) / len(...)` is equivalent to `.mean(axis=1)`

```python
wealth_score = mask_higher.aggregate('sum', axis=1) / len(data.columns)
wealth_score = mask_higher.mean(axis=1)
wealth_score.head()
```

### Group contries by wealth_score and sum values within that bin

```python
binsum_gdp_above_average = data.groupby(wealth_score).sum()
binsum_gdp_above_average
```

```python
binsum_gdp_above_average.to_csv('processed.csv')
```

```python
data = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
data
```

```python
years = data.columns
gdp_australia = data.loc['Australia']
plt.plot(years, gdp_australia, 'g--')
```

```python
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 100
```

```python
gdp_nz = data.loc['New Zealand']
plt.plot(years, gdp_australia, 'b-', label='Australia')
plt.plot(years, gdp_nz, 'g-', label='New Zealand')

plt.legend(loc='lower right')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')
plt.savefig('my_figure.png')
```
